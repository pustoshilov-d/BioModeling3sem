```python
mu = 1
beta = 2
gamma = 2
N = 10

S = lambda s,i: mu*N - beta*s*i - mu*s
I = lambda s,i: beta*s*i - gamma*i - mu-i

fun = lambda p: [ S(p[0],p[1]), I(p[0],p[1]) ]

size = 12
# visualize the vector field
ss = np.linspace(0,size,20)
ii = np.linspace(0,size,20)

Sm,Im = np.meshgrid(ss,ii)

plt.figure(figsize=(10,10))

plt.xlim(0,size);plt.ylim(0,size)

plt.quiver(Sm,Im, 
    S(Sm,Im)/np.sqrt(S(Sm,Im)**2+I(Sm,Im)**2), 
    I(Sm,Im)/np.sqrt(S(Sm,Im)**2+I(Sm,Im)**2), 
    scale_units='xy', angles='xy')
  

plt.plot(N,0, 'ro')
plt.plot((gamma+1)/beta, -(1/beta)+N/(gamma+1), 'ro')

```
