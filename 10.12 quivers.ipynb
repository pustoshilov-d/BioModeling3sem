```python
%matplotlib inline
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 0.7
r = 0.2
K = 2

P = lambda x,y: -r*x - x*r*x/K + a*x*y 
Q = lambda x,y: -r*y - y*r*y/K + a*x*y 

fun = lambda p,t: [ P(p[0],p[1]), Q(p[0],p[1]) ]
```

```python
size = 1
# visualize the vector field
xx = np.linspace(0,size,20)
yy = np.linspace(0,size,20)

Xm,Ym = np.meshgrid(xx,yy)

plt.figure(figsize=(7,7))

plt.quiver(Xm,Ym, 
    P(Xm,Ym)/np.sqrt(P(Xm,Ym)**2+Q(Xm,Ym)**2), 
    Q(Xm,Ym)/np.sqrt(P(Xm,Ym)**2+Q(Xm,Ym)**2), 
    scale_units='xy', angles='xy')

# # plot few phase trajectorie

# drawing stationary points
# plt.plot(0,0, 'ro')
# plt.plot(0, K, 'ro')
# plt.plot(K, 0, 'ro')
# plt.plot(r*K/(r+a*K), r*K/(r+a*K), 'ro')
# plt.xlim(0,size);plt.ylim(0,size);
```
