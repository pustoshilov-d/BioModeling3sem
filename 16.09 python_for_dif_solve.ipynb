```python
import scipy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def pend(y, t, r, k):
    dydt = r*y*(1-y/k)
    return dydt
```

```python
y0=[5.0]
t=np.linspace(0,10,101)
r=1.0
k=10.0
```

```python
from scipy.integrate import odeint
sol = odeint(lambda y,t,r,k: r*y*(1-y/k),y0,t,args=(r,k))
sol[:10,0]
```

```python
plt.plot(t,sol[:,0], 'b', label='y(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
